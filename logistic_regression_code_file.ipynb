Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
# defined new dataset with 492 fraud and 492 (random) legit data points
new_dataset = pd.concat([legit_sample, fraud], axis=0) # axis = 0 means row-wise operation, 1 means column-wise operation
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1) # dropped the 'Class' Column from the dataset
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)  # 0 is Normal, 1 is Fraudulent
```

Split the data into Training data & Testing Data

```python
'''Generally this splitting method is used in Supervised ML, where we know the labelled Y-data (true values we want in predictions)
For accuracy, we match the predited values with these Y labels both in train as well as test datapoints'''
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

Evaluate precision, recall and F1-score: critical metrics in fraud detection.
- Precision reflects the proportion of detected frauds that are true frauds, i.e. “When the model says Positive, how often is it correct?”
- Recall reflects the proportion of actual frauds detected, i.e. “Out of all actual Positives, how many did the model find?”
- F1-score balances both: It is a way to measure a model’s performance by combining precision and recall into a single number.


```python
precision_train = precision_score(Y_train, X_train_prediction)
recall_train = recall_score(Y_train, X_train_prediction)

precision_test = precision_score(Y_test, X_test_prediction)
recall_test = recall_score(Y_test, X_test_prediction)

print(f"Precision_test: {precision_test:.3f} ({precision_test*100:.1f}%)")
print(f"Recall_test:    {recall_test:.3f} ({recall_test*100:.1f}%)")

print(f"Precision_train: {precision_train:.3f} ({precision_train*100:.1f}%)")
print(f"Recall_train:    {recall_train:.3f} ({recall_train*100:.1f}%)")

print("\nTest-Classification Report:")
print(classification_report(Y_test, X_test_prediction, target_names=['Normal', 'Fraud']))

print("\nTrain-Classification Report:")
print(classification_report(Y_train, X_train_prediction, target_names=['Normal', 'Fraud']))

print("\nF1-score results:")
f1_score_test = 2 * (precision_test * recall_test) / (precision_test + recall_test)
print(f"F1 Score_test: {f1_score_test:.3f}")
f1_score_train = 2 * (precision_train * recall_train) / (precision_train + recall_train)
print(f"F1 Score_train: {f1_score_train:.3f}")
```

FEATURE IMPORTANCE ANALYSIS:
- Analyze feature importance based on logistic regression coefficients.
- Highlights which PCA components are most indicative of fraud.


```python
print("\n" + "="*60)
print("FEATURE IMPORTANCE ANALYSIS")
print("="*60)

feature_names = X.columns
coefficients = model.coef_[0]
feature_importance = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients,
    'Abs_Coefficient': np.abs(coefficients),
}).sort_values('Abs_Coefficient', ascending=False).reset_index(drop=True)

print("Top 10 Most Influential Features:")
for i in range(10):
    feature = feature_importance.iloc[i]
    print(f"{i+1:2d}. {feature['Feature']}: {feature['Coefficient']:+.4f}, i.e.{feature['Abs_Coefficient']:+.4f} in absolute")
```
