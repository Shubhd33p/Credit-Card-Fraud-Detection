```python
# =============================================================================
# CREDIT CARD FRAUD DETECTION - END-TO-END PIPELINE
# =============================================================================
# This project demonstrates a complete machine learning pipeline for detecting
# fraudulent credit card transactions using logistic regression.
# =============================================================================

# STEP 1: IMPORTING REQUIRED LIBRARIES
# =============================================================================
"""
Import essential libraries for data manipulation, model building, and evaluation.
"""
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# STEP 2: DATA LOADING AND EXPLORATION
# =============================================================================
"""
Load and explore the credit card transaction dataset.
The dataset contains anonymized transaction data using PCA features V1-V28.
"""
print("Loading Credit Card Fraud Detection Dataset...")
credit_card_data = pd.read_csv("C:\\Users\\Asus\\Downloads\\End2End project\\Credit_Card_Fraud_Detection\\archive\\creditcard.csv")

# Dataset overview
print("\n" + "="*60)
print("DATASET OVERVIEW")
print("="*60)

print(f"Number of transactions: {credit_card_data.shape[0]:,}")
print(f"Number of features: {credit_card_data.shape[1] - 1}")
print(f"Memory usage: {credit_card_data.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Structure and class distribution
print("\nDataset Structure:")
print(credit_card_data.info())

print("\nClass Distribution:")
class_counts = credit_card_data['Class'].value_counts()
print(class_counts)
print(f"Normal transactions: {class_counts[0]:,} ({class_counts[0]/len(credit_card_data)*100:.2f}%)")
print(f"Fraudulent transactions: {class_counts[1]:,} ({class_counts[1]/len(credit_card_data)*100:.2f}%)")

# Data sample and summary
print("\nSample Records:")
print(credit_card_data.head())

print("\nStatistical Summary:")
print(credit_card_data.describe())

# STEP 3: DATA PREPROCESSING
# =============================================================================
"""
Prepare the dataset for training:
- Separate features and target
- Create train/test split with class stratification
"""
print("\n" + "="*60)
print("DATA PREPROCESSING")
print("="*60)

X = credit_card_data.drop(['Class'], axis=1)
y = credit_card_data['Class']

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Feature columns: {list(X.columns)}")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    stratify=y, 
    random_state=2
)

print(f"\nTraining set size: {X_train.shape[0]:,}")
print(f"Testing set size: {X_test.shape[0]:,}")
print(f"Training fraud cases: {y_train.sum():,}")
print(f"Testing fraud cases: {y_test.sum():,}")

# STEP 4: MODEL TRAINING
# =============================================================================
"""
Train a logistic regression model for binary classification.
Chosen for its simplicity, interpretability, and effectiveness.
"""
print("\n" + "="*60)
print("MODEL TRAINING")
print("="*60)

model = LogisticRegression(max_iter=1000, random_state=2)
model.fit(X_train, y_train)

print("Model training completed.")

# Prediction
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# STEP 5: MODEL EVALUATION - ACCURACY
# =============================================================================
"""
Evaluate model accuracy as the overall classification correctness.
"""
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy: {accuracy:.3f} ({accuracy*100:.1f}%)")

# STEP 6: PRECISION AND RECALL EVALUATION
# =============================================================================
"""
Evaluate precision and recall, critical metrics in fraud detection.
- Precision reflects the proportion of detected frauds that are true frauds.
- Recall reflects the proportion of actual frauds detected.
"""
print("\n" + "="*60)
print("PRECISION & RECALL ANALYSIS")
print("="*60)

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(f"Precision: {precision:.3f} ({precision*100:.1f}%)")
print(f"Recall:    {recall:.3f} ({recall*100:.1f}%)")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Normal', 'Fraud']))

# STEP 7: CONFUSION MATRIX AND FALSE POSITIVE ANALYSIS
# =============================================================================
"""
Analyze confusion matrix and false positive rate to assess business impact.
"""
print("\n" + "="*60)
print("CONFUSION MATRIX & FALSE POSITIVE ANALYSIS")
print("="*60)

cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()

print("Confusion Matrix Breakdown:")
print(f"True Negatives (TN): {tn:,}")
print(f"False Positives (FP): {fp:,}")
print(f"False Negatives (FN): {fn:,}")
print(f"True Positives (TP): {tp:,}")

false_positive_rate = fp / (fp + tn)
print(f"\nFalse Positive Rate: {false_positive_rate:.4f} ({false_positive_rate*100:.2f}%)")

# Compare with baseline
baseline_fpr = 0.0035  # Example baseline
current_fpr = false_positive_rate
fp_reduction = ((baseline_fpr - current_fpr) / baseline_fpr) * 100

print("\nFalse Positive Reduction Compared to Baseline:")
print(f"Baseline FP Rate: {baseline_fpr:.4f}")
print(f"Current FP Rate:  {current_fpr:.4f}")
print(f"Improvement: {fp_reduction:.1f}% reduction")

# STEP 8: ADDITIONAL EVALUATION METRICS
# =============================================================================
"""
Compute additional performance metrics:
- Specificity (true negative rate)
- Sensitivity (recall / true positive rate)
- F1 Score (balance between precision and recall)
"""
print("\n" + "="*60)
print("ADDITIONAL PERFORMANCE METRICS")
print("="*60)

specificity = tn / (tn + fp)
sensitivity = tp / (tp + fn)
f1_score = 2 * (precision * recall) / (precision + recall)

print(f"Specificity: {specificity:.3f}")
print(f"Sensitivity (Recall): {sensitivity:.3f}")
print(f"F1 Score: {f1_score:.3f}")

# STEP 9: FEATURE IMPORTANCE ANALYSIS
# =============================================================================
"""
Analyze feature importance based on logistic regression coefficients.
Highlights which PCA components are most indicative of fraud.
"""
print("\n" + "="*60)
print("FEATURE IMPORTANCE ANALYSIS")
print("="*60)

feature_names = X.columns
coefficients = model.coef_[0]
feature_importance = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients,
    'Abs_Coefficient': np.abs(coefficients),
}).sort_values('Abs_Coefficient', ascending=False).reset_index(drop=True)

print("Top 10 Most Influential Features:")
for i in range(10):
    feature = feature_importance.iloc[i]
    print(f"{i+1:2d}. {feature['Feature']}: {feature['Coefficient']:+.4f}")

# STEP 10: PROJECT SUMMARY
# =============================================================================
"""
Summary of model performance and evaluation outcomes.
Includes core metrics relevant to fraud detection effectiveness and business impact.
"""
print("\n" + "="*70)
print("FINAL PROJECT SUMMARY")
print("="*70)

print(f"Dataset size: {credit_card_data.shape[0]:,} transactions")
print(f"Model type: Logistic Regression")
print(f"Accuracy: {accuracy*100:.1f}%")
print(f"Precision: {precision*100:.1f}%")
print(f"Recall: {recall*100:.1f}%")
print(f"Number of PCA features: {X.shape[1]}")
print(f"False positive rate: {current_fpr*100:.2f}%")
print(f"False positive reduction from baseline: {fp_reduction:.1f}%")

print("\nProject execution completed. Model ready for deployment or further tuning.")
```
